```python
#  BATSMAN ARE PLAY MORE BALLS
```

```python
b = df.groupby(['bowler'], as_index=False)['ball'].sum().sort_values(by='ball',ascending=False).head(10)
sns.set(rc={  'figure.figsize':(25,10)})
sns.barplot(x = 'bowler' , y = 'ball' , data = b)
```
